```python
import os
import json
import yaml
# Đường dẫn file annotation và folder output
json_file = r"D:\Downloads\PyTorch\tt100kv1\images\annotations_all.json"  # Đường dẫn đến file JSON
file_yaml = r"D:\Downloads\PyTorch\tt100kv1\dataset.yaml"
output_folder = r"D:\Downloads\PyTorch\tt100kv1\labels"  # Thư mục lưu file TXT

# Tạo thư mục đầu ra nếu chưa tồn tại
os.makedirs(output_folder, exist_ok=True)
```

```python
def change_format_dataset(json_file, file_yaml, output_folder):
    # Đọc file JSON
    with open(json_file, "r") as f:
        data = json.load(f)
    # Mapping từ category sang class_id (bạn cần định nghĩa hoặc lấy từ JSON)
    with open(file_yaml, "r") as file:
        dataset = yaml.safe_load(file)
    category_to_id = {dataset["names"][i]: i for i in range(len(dataset["names"]))}

    # Duyệt qua từng ảnh trong JSON
    for img_id, img_data in data["imgs"].items():
        img_path = img_data["path"]
        objects = img_data.get("objects", [])

        # Bỏ qua ảnh nếu không có đối tượng
        if not objects:
            continue

        # Lấy kích thước ảnh (nếu cần kiểm tra, bạn cần bổ sung width/height từ file JSON)
        # width, height = img_data["width"], img_data["height"]
        width, height = 2048, 2048  # Ví dụ: Kích thước cố định (cần sửa nếu không đồng nhất)

        # Tên file TXT tương ứng với ảnh
        txt_filename = os.path.splitext(os.path.basename(img_path))[0] + ".txt"
        txt_path = os.path.join(output_folder, os.path.dirname(img_path), txt_filename)
        os.makedirs(os.path.dirname(txt_path), exist_ok=True)
        # Ghi dữ liệu vào file TXT
        with open(txt_path, "w") as txt_file:
            for obj in objects:
                bbox = obj["bbox"]
                category = obj["category"]
                if category.find("pa") != -1:
                    print(category)
                    category = "pa*"
                elif category.find("pl") != -1:
                    print(category)
                    category = "pl*"
                elif category.find("pm") != -1:
                    print(category)
                    category = "pm*"
                elif category.find("pr") != -1:
                    print(category)
                    category = "pr*"
                elif category.find("pw") != -1:
                    print(category)
                    category = "pw*"
                elif category.find("ph") != -1:
                    print(category)
                    category = "ph*"
                elif category.find("il") != -1:
                    print(category)
                    category = "il*"
                # Chuyển đổi bbox từ xmin, ymin, xmax, ymax sang x_center, y_center, width, height
                xmin, ymin = bbox["xmin"], bbox["ymin"]
                xmax, ymax = bbox["xmax"], bbox["ymax"]
                x_center = ((xmin + xmax) / 2) / width
                y_center = ((ymin + ymax) / 2) / height
                box_width = (xmax - xmin) / width
                box_height = (ymax - ymin) / height

                # Lấy class_id từ category
                if category not in category_to_id:
                    print(f"Warning: Unknown category '{category}' in image {img_path}")
                    continue
                class_id = category_to_id[category]

                # Ghi vào file TXT theo định dạng YOLO
                txt_file.write(f"{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}\n")
```

```python
change_format_dataset(json_file, file_yaml, output_folder)
```
